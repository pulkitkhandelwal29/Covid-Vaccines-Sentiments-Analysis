```python
# Retrieving the Text Column from the Tweets

covaxin_january_tweets = covaxin_january['Tweet']
covaxin_february_tweets = covaxin_february['Tweet']
covaxin_march_tweets = covaxin_march['Tweet']
covaxin_april_tweets = covaxin_april['Tweet']
covaxin_may_tweets = covaxin_may['Tweet']
covaxin_june_tweets = covaxin_june['Tweet']
covaxin_july_tweets = covaxin_july['Tweet']
covaxin_august_tweets = covaxin_august['Tweet']
covaxin_september_tweets = covaxin_september['Tweet']
```

```python
def clean_tweet(tweets):
    '''
    Helps in cleaning text by removing Tweets URL and @username , removing all characters except Alphabets/Numbers
    and Returns clean Tweets list
    '''
    tweet_clean=[]
    for i in range(len(tweets)):
        #Removing ',' with no space (done for 3,700)
        if re.search(",", tweets[i]):
            result = re.sub('[,]','',tweets[i])
        else:
            result = tweets[i]
        #Replacing https with empty string
        result = re.sub(r"http\S+","",result)
        #Replacing @username with empty string
        result = re.sub(r"@\S+","", result)
        #Replacing all with empty space except Alphabets and digits
        result = re.sub('[^a-zA-Z\d]',' ',result)

        tweet_clean.append(result)
    return tweet_clean
```

```python
#Getting clean tweets 

covaxin_january_clean_tweets = clean_tweet(covaxin_january_tweets)
covaxin_february_clean_tweets = clean_tweet(covaxin_february_tweets)
covaxin_march_clean_tweets = clean_tweet(covaxin_march_tweets)
covaxin_april_clean_tweets = clean_tweet(covaxin_april_tweets)
covaxin_may_clean_tweets = clean_tweet(covaxin_may_tweets)
covaxin_june_clean_tweets = clean_tweet(covaxin_june_tweets)
covaxin_july_clean_tweets = clean_tweet(covaxin_july_tweets)
covaxin_august_clean_tweets = clean_tweet(covaxin_august_tweets)
covaxin_september_clean_tweets = clean_tweet(covaxin_september_tweets)
```
